# Menemukan Outlier dengan Aturan 3-Sigma

```python
mean = df_noid[['goals']].mean()
stdev = df_noid[['goals']].std()
iso = (df_noid[['goals']] < mean - 3*stdev) \
      | (df_noid[['goals']] > mean + 3*stdev)
df1 = df_noid[['player_name', 'goals']].assign(is_outlier = iso)
df1.loc[df1['is_outlier']]
```

# Menemukan Outlier with Tukey's fences

```python
q1 = df_noid['goals'].quantile(q = 0.25)
q3 = df_noid['goals'].quantile(q = 0.75)
iqr = q3 - q1
iso = (df_noid['goals'] < q1 - 1.5*iqr) | (df_noid['goals'] > q3 + 1.5*iqr)
iseo = (df_noid['goals'] < q1 - 3*iqr) | (df_noid['goals'] > q3 + 3*iqr)
df1 = df_noid[['player_name', 'goals']].assign(is_outlier = iso, is_extreme_outlier = iseo)
df1.loc[df1['is_outlier'] | df1['is_extreme_outlier']]
```

# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data yang sudah disesuaikan (14 negara, 14 populasi)
countries = ('Argentina', 'Bolivia', 'Brazil', 'Chile', 'Columbia', 'Ecuador',
             'Falkland Islands', 'French Guiana', 'Guyana', 'Paraguay',
             'Peru', 'Suriname', 'Uruguay', 'Venezuela')
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409, 7107305, 32880332, 585169,
               3470475, 28258770)

df = pd.DataFrame({
    'Country': countries,
    'Population': populations
})
df = df.sort_values(by='Population', ascending=True)

x_coords = np.arange(len(df))
colors = ['#0000FF' for _ in range(len(df))]
colors[-2] = '#FF0000'

plt.figure(figsize=(20,10))
plt.bar(x_coords, df['Population'], tick_label=df['Country'], color=colors)
plt.xticks(rotation=90)  # Nama fungsi dan sintaks parameter sudah diperbaiki
plt.ylabel('Population (Millions)')
plt.title('South American Populations')
plt.show()
```

# Heatmap (peta panas)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mendefinisikan Kota
cities = ['New York', 'Beijing', 'Tokyo', 'Osaka', 'Shanghai',
          'Cairo', 'Delhi', 'Karachi', 'Dhaka', 'Mexico City', 'Mumbai',
          'Sao Paulo']

# mendefinisi bulan-bulan dalam setahun
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

temperatures = [
    [ 4,  6, 11, 18, 22, 27, 29, 29, 25, 18, 13,  7], # New York
    [ 2,  5, 12, 21, 27, 30, 31, 30, 26, 19, 10,  4], # Beijing
    [10, 10, 14, 19, 23, 26, 30, 31, 27, 22, 17, 12], # Tokyo
    [ 9, 10, 14, 20, 25, 28, 32, 33, 29, 23, 18, 12], # Osaka
    [ 8, 10, 14, 20, 24, 28, 32, 32, 27, 23, 17, 11], # Shanghai
    [19, 21, 24, 29, 33, 35, 35, 35, 34, 30, 25, 21], # Cairo
    [20, 24, 30, 37, 40, 39, 35, 34, 34, 33, 28, 22], # Delhi
    [26, 28, 32, 35, 36, 35, 33, 32, 33, 35, 32, 28], # Karachi
    [25, 29, 32, 33, 33, 32, 32, 32, 32, 31, 29, 26], # Dhaka
    [22, 24, 26, 27, 27, 26, 24, 25, 24, 24, 23, 23], # Mexico City
    [31, 32, 33, 33, 34, 32, 30, 30, 31, 34, 34, 32], # Mumbai
    [29, 29, 28, 27, 23, 23, 23, 25, 25, 26, 27, 28], # Sao Paulo
]

plt.figure(figsize=(12, 8))
sns.heatmap(
    temperatures,
    yticklabels = cities,
    xticklabels = months,  # Sekarang variabel months sudah terdefinisi
    cmap = 'coolwarm',
    annot = True,  # Menampilkan nilai temperatur
    fmt = 'd',  # Format angka sebagai integer
    linewidths = 0.5  # Garis pembatas antar sel
)

plt.title('Average Monthly Temperatures (°C) in Major Cities')
plt.xlabel('Month')
plt.ylabel('City')
plt.tight_layout()  # Agar label tidak terpotong
plt.show()
```

# Grouping (pengelompokan)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Pertama buat tabel pivot (contoh menggunakan data mobil)
# Dengan asumsi df berisi kolom: 'drive-wheels', 'body-style', 'price'
grouped = df.groupby(['drive-wheels', 'body-style'])['price'].mean()
grouped_pivot = grouped.unstack()  # Konversi ke format tabel pivot

# Buat plotnya
fig, ax = plt.subplots(figsize=(10, 8))

# Create heatmap
im = ax.pcolor(grouped_pivot,
               cmap='RdBu',
               edgecolors='w',  # white grid lines
               linewidths=0.5)  # line width

# Tetapkan label - versi yang disederhanakan
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5)
ax.set_xticklabels(grouped_pivot.columns, rotation=90)
ax.set_yticklabels(grouped_pivot.index)

# Add colorbar
cbar = fig.colorbar(im)
cbar.set_label('Price ($)', rotation=270, labelpad=20)

# Add titles
ax.set_title('Average Price by Drive Wheels and Body Style', pad=20)
ax.set_xlabel('Body Style')
ax.set_ylabel('Drive Wheels')

# Improve layout
plt.tight_layout()
plt.show()
```
